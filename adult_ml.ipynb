```python
import pandas as pd
import numpy as np
```

Read dataset

```python
df = pd.read_csv('adult.csv')
df.head()
```

preprocessing 

```python
# check if there the nan values
df.isna().sum()
```

```python

df.isin(['?']).sum()
```

Replace '?' values with 'nan' 

```python
df['workclass']=df['workclass'].replace({'?':np.NaN})
df['occupation']=df['occupation'].replace({'?':np.NaN})
df['native-country']=df['native-country'].replace({'?':np.NaN})
df.isin(['?']).sum()
```

Drop the nan values

```python
df.dropna(inplace=True)
```

Drop duplication Data

```python
df=df.drop_duplicates()
df
```

```python
df.shape
```

```python
df.describe()
```

Replace values in income col if ">50" with 1 , else 0

```python
df['income'] = [1 if i.strip() == " >50" else 0 for i in df.income]
```

```python
df
```

convert the data to catogrical data

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
df.astype("int")
```

divided the data to train and test

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x  = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=1)
```

Apply The KNN

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn_model = KNeighborsClassifier().fit(x_train, y_train)
y_pred = knn_model.predict(x_test)

print(accuracy_score(y_test, y_pred))
```
